```python
group0 = Project3['TenYearCHD'] == 0
group1 = Project3['TenYearCHD'] == 1
variable = Project3['age']
print(f"var1{variable[group0] .var(): .3f}" \
      f"var2 {variable[group1] .var():03f}")
```

```python
variable = Project3['age']
t, pvalue = ttest_ind(variable[group0], variable[group1],
                               axis = 0, equal_var = False)
print(f"t statistic {t: .3f} p-value {pvalue: .3f}")
```

There is a statistically significant difference in the ages of persons with Chronic Heart Disease and those without Chronic Heart Disease.

```python
X = Project3.loc[:, Project3.columns != 'TenYearCHD']
y = Project3.loc[:, Project3.columns == 'TenYearCHD']
```

```python
import statsmodels.api as sm
logit_model=sm.Logit(y,X)
result=logit_model.fit()
print(result.summary2())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(y_test, logreg.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

Scatterplot: It creates a scatterplot to visually explore the relationship between age and systolic blood pressure.
Boxplot: It creates a boxplot to compare Body Mass Index (BMI) across different age groups. This helps identify trends or differences in BMI based on age.
Heatmap: It calculates the correlation between all the features in the dataframe and displays it as a heatmap. The heatmap shows the strength and direction of the relationships between features.
```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['age'])
plt.show()

sns.scatterplot(x='age', y='sysBP', data= df)
plt.show()

sns.boxplot(x='age', y='BMI', data=df)
plt.xticks(rotation=90)
plt.show()

sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

# Create a Logistic Regression model
model = LogisticRegression()

# Define X_train and y_train
X_train = df[['age', 'sysBP']]
y_train = df['TenYearCHD']
# Train the model
model.fit(X_train, y_train)
```

 test and predict a categorical target variable.

```python
X_test = [[50, 120], [60, 140]]

# Predict the class labels
y_pred = model.predict(X_test)

# Print the predicted class labels
print(y_pred)
```

This code snippet prepares a small test set with two data points and uses a trained model to predict class labels for those data points.

Use appropriate metric and interpret the meaning of the score.

```python
from sklearn.metrics import accuracy_score

y_true = [0, 1, 1, 0, 1]
y_pred = [0, 1, 0, 0, 1]

accuracy = accuracy_score(y_true, y_pred)


# Print the accuracy score
print("Accuracy:", accuracy)

# Interpret the meaning of the score
if accuracy > 0.5:
  print("The model is able to correctly classify more than 50% of the data points.")
else:
  print("The model is not able to correctly classify more than 50% of the data points.")
```

The model achieved an accuracy of 80%, indicating a strong capability to generalize from the training data to unseen examples. This performance suggests that the model has effectively learned the underlying relationships within the data and can be successfully applied to the intended task.
